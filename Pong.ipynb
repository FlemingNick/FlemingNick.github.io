# Pong

### Controlls: 

##### Player 1 (left): w - up, e - down

##### Player 2 (right): o - up, p - down


```python
import turtle
import os #erlaubt das zugreifen auf das system (wird für Sound benötigt)

wn = turtle.Screen() #Screen
wn.title("Pong")
wn.bgcolor("black") #Hintergrundfarbe
wn.setup(width = 800, height = 600)
wn.tracer(0) #Stops window from updating

#Score

score_a = 0
score_b = 0

#Paddle A

paddle_a = turtle.Turtle()
paddle_a.speed(0) #setzt Speed auf max.
paddle_a.shape("square")
paddle_a.color("white")
#Standartshape is 20x20 Pixel
paddle_a.shapesize(stretch_wid=5,stretch_len=1) #Multiplikator des Stretchings
paddle_a.penup() #Turtle zieht Linie bei Bewegung, so wirds ausgeschaltet
paddle_a.goto(-350, 0) #setzt Startkoordinaten


#Paddle B

paddle_b = turtle.Turtle()
paddle_b.speed(0) #setzt Speed auf max.
paddle_b.shape("square")
paddle_b.color("white")
#Standartshape is 20x20 Pixel
paddle_b.shapesize(stretch_wid=5,stretch_len=1) #Multiplikator des Stretchings
paddle_b.penup() #Turtle zieht Linie bei Bewegung, so wirds ausgeschaltet
paddle_b.goto(350, 0)

#Ball

ball = turtle.Turtle()
ball.speed(0) #setzt Speed auf max.
ball.shape("square")
ball.color("white")
ball.penup() #Turtle zieht Linie bei Bewegung, so wirds ausgeschaltet
ball.goto(0, 0)
ball.dx = 2 #dx = delta x = veränderung auf x. Jede Bewegung entspricht 2 Pixeln
ball.dy = 2


#Pen

pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle
pen.goto(0, 260)
pen.write("Player A: 0  Player B: 0", align="Center", font=("Courier", 24, "normal"))
#Zuerst bewegt er sich also 2 Pixel nach rechts und 2 nach oben, da positive Vorzeichen


#Function

def paddle_a_up():
    y = paddle_a.ycor() #returns y-coordinate
    y += 20 #fügt 20 Pixel hinzu
    paddle_a.sety(y) #Ersetzt neue durch alte koordinaten
    #bis jetzt is Funktion nur definiert, muss aber noch abgerufen werden
    
def paddle_a_down():
    y = paddle_a.ycor()
    y-= 20
    paddle_a.sety(y)
    
def paddle_b_up():
    y = paddle_b.ycor()
    y += 20
    paddle_b.sety(y)
    
def paddle_b_down():
    y = paddle_b.ycor()
    y -= 20
    paddle_b.sety(y)

#Keyboard binding

wn.listen() #nun reagiert es auf Inputs
wn.onkeypress(paddle_a_up, "w") #wenn w gedrückt wird, wird paddle_a_up ausgeführt
wn.onkeypress(paddle_a_down, "e")
wn.onkeypress(paddle_b_up, "o")
wn.onkeypress(paddle_b_down, "p")



#Main game loop

while True:
    wn.update()   #refreshed den screen
    
    #Move the bal
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    #Border checking
    
    if ball.ycor() > 290:
        ball.sety(290)
        ball.dy *= -1
        os.system("afplay bounce.wav&") #aplay für Linux, & Zeichen damit es flüssig läuft beim Sound abspielen
    
    if ball.ycor() < -290:
        ball.sety(-290)
        ball.dy *= -1 
        os.system("afplay bounce.wav&")
        
    if ball.xcor() > 390:
        ball.goto(0, 0) #setzt ball wieder zum Center
        ball.dx *= -1 #Dreht Richtung wieder um
        score_a += 1
        pen.clear() #sonst wird neue Zahl nur auf alte Zahl geschrieben
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align="Center", font=("Courier", 24, "normal"))
        
    if ball.xcor() < -390:
        ball.goto(0, 0)
        ball.dx *= -1
        score_b += 1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align="Center", font=("Courier", 24, "normal"))
    
    #Kollision
    if (ball.xcor() > 340 and ball.xcor() < 350) and ball.ycor() < paddle_b.ycor() + 50 and ball.ycor() > paddle_b.ycor() - 50:    #Da Paddle 10 Pixel breit ist und 100 Pixel lang
        ball.setx(340) #sonst gibt es einen Glitch
        ball.dx *= -1
        os.system("afplay bounce.wav&")
    if (ball.xcor() < -340 and ball.xcor() > -350) and ball.ycor() < paddle_a.ycor() + 50 and ball.ycor() > paddle_a.ycor() - 50:    #Da Paddle 10 Pixel breit ist und 100 Pixel lang
        ball.setx(-340) #sonst gibt es einen Glitch
        ball.dx *= -1  
        os.system("afplay bounce.wav&")
    
 
```
